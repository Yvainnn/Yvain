```python
import numpy as np
import matplotlib.pyplot as ple
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
%matplotlib inline

iris=load_iris()
keys_list=iris.keys()
data=iris.data
featn=iris.feature_names

#print("Shape of 'data':", data.shape)
#print("Type of 'data':", type(data))
#print("Type of 'featn':", type(featn))

df=pd.DataFrame(data=data, columns=featn)
#print(df)

summary=df.describe()
#print(summary)

df_subset=df.iloc[:50]
#print(df_subset)

#print(df['sepal length (cm)'].describe())
sepal_l=df_subset['sepal length (cm)']
#print(sepal_l)

mean_sepal=sepal_l.mean()
std_sepal=sepal_l.std()
#print(mean_sepal, std_sepal)

for val in sepal_l: 
    score=(val-mean_sepal)/std_sepal
    if score>3.9: 
        print('Ex. values identified', val)



def find_extreme_values(db_set):
    sepal = df_subset[db_set]
    sepal_mean = sepal.mean()
    sepal_std = sepal.std()
    h=float(input('def score  '))
    for val in sepal:
        score = (val - sepal_mean) / sepal_std
        if score > h:
            print('Extreme value identified:', val)


#user = input('Select one db: sepal length, sepal width, petal length, petal width ')
#find_extreme_values(user)

plt.boxplot(df)
plt.title('Yvain')
plt.xlabel('values')
plt.ylabel('feat')
plt.savefig('iris_feat_distr.png')

df_subset.to_csv('iris_database', index=False)
```
